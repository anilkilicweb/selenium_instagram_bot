```python
import pandas as pd
```

```python
df_followers = pd.read_csv('CSV PATH')
df_following = pd.read_csv('CSV PATH')
```

```python
df_followers = pd.DataFrame(df_followers)
df_following = pd.DataFrame(df_following)
df = pd.concat([df_followers, df_following], axis=1)
df.head()
```

# users not following you back

```python
def highlight_cols(x):
    colored_df_nfb = []
    for i in x:
        if i in set(df['followers']):
            colored_df_nfb.append('background: green')
        else:
            colored_df_nfb.append('background: red')
    return colored_df_nfb

# apply highlighting logic to first column only
colored_df_nfb = df_following.style.apply(highlight_cols, subset=pd.IndexSlice[:, ['following']])
colored_df_nfb
```

```python
df['following'].isin(df['followers']).value_counts() # 29 people are not following you back.
```
