```python
import pandas as pd
from IPython.display import display
# pd.set_option('display.max_rows', None)
```

```python
df_followers = pd.read_csv('followers.csv')
df_following= pd.read_csv('following.csv')
# display(df_followers, df_following)
df = pd.concat([df_followers, df_following], axis=1)
df
```

```python
df['following'].isin(df['followers']).value_counts()
```

# Users not following you back

```python
def highlight_cols(x):
    colored_df_nfb = []
    for i in x:
        if i in set(df['followers']):
            colored_df_nfb.append('background: green')
        else:
            colored_df_nfb.append('background: red')
    return colored_df_nfb

# apply highlighting logic to first column only
colored_df_nfb = df.style.apply(highlight_cols, subset=pd.IndexSlice[:, ['following']])
colored_df_nfb
```

# Users you are not following back

```python
def highlight_cols(x):
    colored_df_ynfb = []
    for i in x:
        if i in set(df['following']):
            colored_df_ynfb.append('background: green')
        else:
            colored_df_ynfb.append('background: red')
    return colored_df_ynfb

# apply highlighting logic to first column only
colored_df_ynfb = df.style.apply(highlight_cols, subset=pd.IndexSlice[:, ['followers']])
colored_df_ynfb
```
